```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print(sns.get_dataset_names())

crashes_df=sns.load_dataset('car_crashes')
crashes_df.head()
```

```python
sns.distplot(crashes_df['not_distracted'], kde=False, bins=25)
```

```python

sns.jointplot(data=crashes_df,x='alcohol',y='not_distracted',kind='reg')
```

```python
sns.kdeplot(crashes_df['alcohol'])
```

```python
sns.pairplot(data=crashes_df)
```

```python
tips_df=sns.load_dataset('tips')
tips_df.head()
```

```python
sns.set_style('darkgrid')
sns.set_context('poster',font_scale=.5)
sns.rugplot(data=tips_df['total_bill'])
```

```python
sns.barplot(data=tips_df,x='sex',y='tip',estimator="median")
```
